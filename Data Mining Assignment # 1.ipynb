## 4. Drop Columns who have more than 50 percent missing values 

```python
colTodel=[]
for i in range(len(DataSet.columns)):
    tempDf=DataSet[pd.isnull(DataSet.iloc[:,i])]
    if(len(tempDf)/len(DataSet)>0.5):
        colTodel.append(DataSet.iloc[:,i].name)
print("these columns needs to be deleted",colTodel)
DataSet.drop(colTodel,inplace=True,axis=1)
        
    
```

## 5. Replace missing values with Median 

```python
RecToreplaceWithMedian=['age','fare']
for i in RecToreplaceWithMedian:
    DataSet[i].fillna((DataSet[i].median()), inplace=True)
print("This is the updated dataset\n",DataSet)
```

## 6. Replace missing address with a default address 

```python
DataSet['home.dest'].fillna('House#10, I-10/2', inplace=True)
DataSet
```

## 7. Replace missing values with Mean

```python
tempReplacedWithMean=DataSet['age'].fillna((DataSet['age'].mean()), inplace=False)
print("Temporarily replace NA with Mean\n",tempReplacedWithMean)
```
